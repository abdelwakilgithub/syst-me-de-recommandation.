```python
import pandas as pd
import numpy as np
```

```python
excel = pd.read_excel("Reviews-Asmi.xlsx")
```

```python
table_rating = excel.rename(columns={'reviewerName': 'Users','categories': 'Products'})
```

```python
table_rating
```

```python
table_rating_new = table_rating.groupby(['Users', 'Products']).agg({'rating': ['mean']})
```

```python
matriceless_rating = pd.pivot_table(table_rating_new,index=["Users"],columns=['Products'], values=['rating'])
```

```python
matriceless_rating
```

```python
matrice_rating = pd.pivot_table(table_rating_new,index=["Users"],columns=['Products'], values=['rating']).fillna(0)
```

```python
matrice_rating
```

```python
from sklearn.metrics.pairwise import cosine_similarity
matrice_similarite = cosine_similarity(matrice_rating, matrice_rating)
```

```python
matrice_similarite
```

```python
multiplication_vector = matrice_rating.dot(matrice_rating.T)
```

```python
norm_user = np.sqrt((matrice_rating**2).sum(axis = 1))
```

```python
matrice_norm = np.outer(norm_user, norm_user)
```

```python
matrice_similarite_cacul = multiplication_vector/matrice_norm
```

```python
matrice_similarite_cacul
```
